2. Count Word Frequency in Customer Reviews
ðŸ’¼ Real-world use: Analyzing reviews or comments in text analytics.
Task:
Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent, awesome"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

```python
sentence = input("Review: ").lower()
words = [w.strip(".,!?\"'") for w in sentence.split()]

unique = []
for w in words:
    if w not in unique:
        unique.append(w)

word_counts = [(w, words.count(w)) for w in unique]  
print({w: c for w, c in word_counts})
```

3. Password Strength Checker
ðŸ’¼ Real-world use: Used in user authentication systems.
Task:
Check if a password is strong â€” must have at least one uppercase, one lowercase, one digit, and one
special character.
Example:
Input: "Hello@123"
Output: Strong Password

```python
password = input("Password: ")
if ( len(password) >= 8 and
     any(c.islower() for c in password) and
     any(c.isupper() for c in password) and
     any(c.isdigit() for c in password) and
     any(c in "!@#$%^&*()-_+=" for c in password) ):
    print("Strong password")
else:
    print("Weak password")
```

5. SMS Word Counter

Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).
Example:

```python
msg = input("Enter your message: ")
char_count = len(msg)
words = len(msg.split())
status = "Within Limit" if char_count <= 160 else "Exceeded Limit"
print(f"Words = {words}, Characters = {char_count}, Status = {status}")
```

6. Remove Duplicates from Contact List
ðŸ’¼ Real-world use: Data cleaning in CRM or user databases.
Task:
Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

```python
names = input("Enter names separated by commas: ")
name_list = [name.strip().lower() for name in names.split(",")]
unc_names = []
for name in name_list:
    if name not in unc_names:
        unc_names.append(name)

print([name.capitalize() for name in unc_names])
```

7. Detect Suspicious Words in Text (Moderation)
ðŸ’¼ Real-world use: Used in chat moderation or email filtering.
Task:
If any â€œbannedâ€ words like ['spam', 'scam', 'fake'] are found in a text, print â€œWarningâ€.
Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected

```python
banned_words = ["spam", "scam", "fake"]
message = input("Enter your message: ").lower()
if any(bw in message for bw in banned_words):
    print("Warning: Suspicious content detected!")
else:
    print("Message is clean.")  
```

9. Count Hashtags and Mentions (Social Media Analytics)
ðŸ’¼ Real-world use: Used in analyzing social media posts.
Task: 
Count how many hashtags (#) and mentions (@) appear in a post.
Example: 
Input: "Check out #AIWorkshop and follow @inceptionbd"
Output: Hashtags = 1, Mentions = 1

```python
sen = input("Enter your post: ").lower()
sen = sen.split(" ")
hashtags = 0
for word in sen:
    if word.startswith("#") and word[1:].isalpha():
        hashtags += 1
mentions = 0
for word in sen:
    if word.startswith("@") and word[1:].isalpha():
        mentions += 1
print(f"Hashtags = {hashtags}, Mentions = {mentions}")
```

10. Word Reversal for Data Encryption
ðŸ’¼ Real-world use: Simple text obfuscation or playful encoding.
Task:
Reverse each word in a sentence while keeping word order the same.
Example:
Input: "AI is fun"
Output: "IA si nuf"

```python
sen = input("Enter your post: ")
sen = sen.split(" ")

for i in range(len(sen)):
    print(sen[i][::-1], end=" ")
```

11.
Write a program that calculates a user's total monthly electricity bill based on the number of units consumed. The calculation must follow a tiered rate structure:
â—The first 100 units are billed at 5 BDT per unit.
â—The next 100 units are billed at 7 BDT per unit.
â—All next units above 200 are billed at 10 BDT per unit. 

```python
unit = int(input("Enter number of units consumed: "))
if unit <= 100:
    bill = unit * 5 
elif unit <= 200:
    bill = 100 * 5 + (unit - 100) * 7
else:
    bill = 100 * 5 + 100 * 7 + (unit - 200) * 10
    
print(f"Total bill amount: {bill}")
```
