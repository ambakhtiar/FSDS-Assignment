Assignment 1: Temperature Check

```python
temp = float(input("Enter Temperature: "))
if temp <=10 :
    print("It's Cold")
elif temp > 10 and temp <=25 :
    print("It's cool outside")
else:
    print("It's Hot outside")
```

Assignment 2: ATM Withdrawal System

```python
balance = int(input("Enter account balance: "))
withdraw = int(input("Enter amount to withdraw: "))
if withdraw <= balance:
    print(f"Transaction successful.")
else:
    print("Insufficient balance.")
```

Assignment 3: Student Grade Calculator

```python
marks = int(input("Enter your marks: "))
if marks >= 90:
    grade = 'A+'
elif marks >= 80:
    grade = 'A'
elif marks >= 70:
    grade = 'B'
elif marks >= 60:
    grade = 'C'
else:
    grade = 'Fail'
print(f"Your grade is: {grade}")
```

Assignment 4: Bus Ticket Discount

```python
age = int(input("Enter your age: "))
ticketPrice = 100
if age < 5:
    ticketPrice = 0
elif age >= 5 and age <= 18:
    ticketPrice *= 0.5
elif age >= 60:
    ticketPrice *= 0.7
print(f"Your 100tk ticket price is: {ticketPrice}")
```

Assignment 5: Restaurant Bill with Discount

```python
bill = float(input("Enter your bill amount: "))
discount = 0
if bill >= 1000:
    discount = bill * 0.1
    bill -= discount
print(f"Discount Amount: {discount}tk. \nYour final bill amount is: {bill}tk.")
```

Assignment 7: Traffic Light System

```python
traffic_light = input("Enter traffic light color (red/yellow/green): ").lower()
if traffic_light == "red":
    print("Stop")
elif traffic_light == "yellow":
    print("Ready to go")    
elif traffic_light == "green":
    print("Go")
else:
    print("Invalid color")
```

Assignment 8: Leap Year Checker

```python
year = int(input("Enter a year: "))
if year %4==0 and (year %100 !=0 or year %400==0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

Assignment 10: Electricity Bill Calculator

```python
units = int(input("Enter number of electric units consumed: "))
bill = 0
if units <= 100:
    bill = units * 5
elif units >100 and units<=200:
    bill = (100 * 5) + ((units-100)*7)
else:
    bill = (100 * 5) + ((units-100)*7) + ((units-200)*10)

print(f"Your usage units {units} \nTotal bill {bill}")

```
